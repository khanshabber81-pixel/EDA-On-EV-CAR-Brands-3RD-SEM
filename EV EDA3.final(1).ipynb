## Step 3: Data Cleaning

### Why Data Cleaning?
- **Remove Duplicates**: Duplicate rows can skew analysis and introduce bias
- **Handle Missing Values**: Missing data can lead to inaccurate results and model errors
- **Ensure Data Quality**: Clean data improves reliability of insights and predictions

### Conditions for Our Cleaning Steps:

**1. Remove Duplicates - When?**
   - Condition: When the same record appears more than once in the dataset
   - Reason: Duplicates artificially inflate counts and bias statistical calculations
   - Example: If the same electric vehicle model is recorded twice, it counts as 2 instead of 1

**2. Drop Rows with Missing Values - When?**
   - Condition: When any column in a row has NULL/NaN values
   - Reason: Missing values can break analysis and machine learning models
   - Example: If a vehicle record is missing the "Price" field, we can't analyze pricing accurately with that row

```python
# Remove duplicate rows
df = df.drop_duplicates()
df.shape
```

**Remove Duplicates**: Check for and eliminate duplicate rows that can distort statistical analysis

### 5.2 Create Filtered Dataset for Top 10 Brands

```python
# Create subset with top 10 brands only
top_10_brand_names = top_10_brands.index.tolist()
df_top_10 = df[df[brand_column].isin(top_10_brand_names)].copy()

print(f"Original dataset: {len(df)} rows")
print(f"Top 10 brands dataset: {len(df_top_10)} rows")
print(f"Percentage of data: {(len(df_top_10)/len(df)*100):.1f}%")
```

## Step 6: Detailed EDA - Univariate, Bivariate & Multivariate Analysis

### 6.1 Univariate Analysis - Distribution of Key Features

**Insight: Most EVs cluster around 60-80 kWh batteries and 350-500km range**

```python
# Get numeric columns
num_features = df_top_10.select_dtypes(include=[np.number]).columns.tolist()[:4]

fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Univariate Analysis - Distribution of Key Features (Top 10 Brands)', 
             fontsize=16, fontweight='bold', y=0.995)

for idx, col in enumerate(num_features):
    row = idx // 2
    col_idx = idx % 2
    
    axes[row, col_idx].hist(df_top_10[col], bins=30, color='steelblue', edgecolor='black', alpha=0.7)
    axes[row, col_idx].set_title(f'Distribution of {col}', fontsize=13, fontweight='bold')
    axes[row, col_idx].set_xlabel(col, fontsize=11)
    axes[row, col_idx].set_ylabel('Frequency', fontsize=11)
    axes[row, col_idx].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```

### 6.2 Bivariate Analysis - Relationships Between Variables

**Insight 1: Strong positive correlation between battery size and range across all brands**

```python
# Battery vs Range analysis
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

# Scatter with brand colors
for brand in df_top_10[brand_column].unique():
    brand_data = df_top_10[df_top_10[brand_column] == brand]
    axes[0].scatter(brand_data[num_features[0]], brand_data[num_features[1]], 
                   label=brand, alpha=0.6, s=120, edgecolor='black', linewidth=0.5)

axes[0].set_xlabel(f'{num_features[0]} (Range)', fontsize=12, fontweight='bold')
axes[0].set_ylabel(f'{num_features[1]} (Battery Size)', fontsize=12, fontweight='bold')
axes[0].set_title('Battery Size vs Range - Positive Correlation Across Brands', fontsize=13, fontweight='bold')
axes[0].legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)
axes[0].grid(True, alpha=0.3)

# Hexbin density plot
hb = axes[1].hexbin(df_top_10[num_features[0]], df_top_10[num_features[1]], 
                     gridsize=20, cmap='YlOrRd', mincnt=1)
axes[1].set_xlabel(f'{num_features[0]} (Range)', fontsize=12, fontweight='bold')
axes[1].set_ylabel(f'{num_features[1]} (Battery Size)', fontsize=12, fontweight='bold')
axes[1].set_title('Density Distribution - Battery vs Range Clusters', fontsize=13, fontweight='bold')
cbar = plt.colorbar(hb, ax=axes[1], label='Count')

plt.tight_layout()
plt.show()
```

**Mercedes-Benz dominates the high-performance segment with the largest batteries and longest ranges, while brands like BYD and Kia compete effectively in the mid-range efficiency sweet spot.**

```python
# Efficiency analysis
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

# Correlation scatter
col1, col2 = num_features[1], num_features[2]
axes[0].scatter(df_top_10[col1], df_top_10[col2], alpha=0.6, s=120, 
               color='coral', edgecolor='darkred', linewidth=0.5)
axes[0].set_xlabel(f'{col1}', fontsize=12, fontweight='bold')
axes[0].set_ylabel(f'{col2}', fontsize=12, fontweight='bold')
axes[0].set_title(f'{col1} vs {col2} Relationship', fontsize=13, fontweight='bold')
axes[0].grid(True, alpha=0.3)

# Correlation heatmap
correlation_matrix = df_top_10[num_features].corr()
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', 
            center=0, ax=axes[1], cbar_kws={'label': 'Correlation'}, 
            square=True, linewidths=1.5, annot_kws={'size': 11})
axes[1].set_title('Correlation Matrix - Key Features', fontsize=13, fontweight='bold')

plt.tight_layout()
plt.show()
```

**Battery capacity, top speed, and torque are all strongly interconnected (0.74-0.77 correlations), while the number of battery cells shows a weaker relationship, suggesting that cell configuration matters less than total capacity for performance.**
